```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
diamond = sns.load_dataset('diamonds')
```

```python
diamond
```

```python
diamond.head()
```

**Point Plots**

```python
sns.distplot(diamond['price'], kde=False)
```

```python
sns.kdeplot(diamond['price'])
```

**Distribution Plots**

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips.head()
```

```python
sns.residplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='sex')
```

```python
x = sns.pairplot(tips)
x.map_diag(sns.distplot)
```

```python
x.map_upper(sns.kdeplot)
```

**Categorical Plots**

```python
sns.pointplot(x='day', y='total_bill', data=tips)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, palette='cool')
```

**Matrix Plots**

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
tips.head()
```

```python
tip = tips.corr()
```

```python
tip
```

```python
sns.heatmap(tip, annot=True, linewidths=3)
```

```python
flight = flights.pivot(index='month', columns='year', values='passengers')
```

```python
flight
```

```python
sns.heatmap(flight, linewidths=1)
```

```python
sns.clustermap(flight, cmap='CMRmap')
```
